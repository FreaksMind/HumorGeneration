```python
import pandas as pd
import re
```

```python
input = pd.read_csv("data/task-a-en.tsv", sep="\t")
results = pd.read_csv("results/task-a-en.tsv", sep="\t")
```

```python
word_based_ids = input[
    input["headline"] == "-"
]["id"]
```

```python
print(len(word_based_ids))
```

```python
print(word_based_ids)
```

```python
news_only = results[
    ~results["id"].isin(word_based_ids)
]
```

```python
len(news_only)
```

```python
META_PATTERNS = [
    r"\(get it\?\)",
    r"\(note:.*?\)",
    r"\(source:.*?\)",
    r"\(inspired by.*?\)",
    r"\(sorry.*?\)",
    r"\(i hope.*?\)",
    r"\(answer.*?\)"
]
```

```python
def clean_joke(text):
    text = text.strip()

    text = re.sub(r"^(this is a joke:|no need to explain:|answer:|answer should be|the joke |)\s*", "", text.lower(), flags=re.I)

    for pattern in META_PATTERNS:
        match = re.search(pattern, text.lower(), flags=re.I)
        if match:
            text = text[:match.start()]
            break

    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
news_only["text"] = news_only["text"].apply(clean_joke)
```

```python
news_only.to_csv(
    "results/task-a-en.tsv",
    sep="\t",
    index=False
)
```
